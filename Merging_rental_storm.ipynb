```python
#import the necessary packaged for managing datasets
import pandas as pd
import datetime
from datetime import datetime as dt
```

```python
# Read the CSV  file containing all the datafields
a = pd.read_csv("pidfwithlatlong.csv",names = ['trip_id', 'start_time', 'stop_time', 'bike_id', 'trip_duration','from_station_id', 'from_station_name', 'to_station_id','station_name', 'user_type  ', 'from_latitude', 'from_longitude','to_latitude', 'to_longitude', 'DATE', 'AWND', 'PRCP', 'TMAX', 'TMIN', 'WT01', 'WT02', 'WT03', 'WT04', 'WT08', 'WT09', 'season', 'holiday','walk_score', 'transit_score', 'bike_score', 'distance'] ,header = 0)
```

```python
# Display first 3 observations
a.head(3)
```

```python
## Display the columns
a.columns
```

```python
#Rounding off the distance from original station to destination station to 2 decimal places after decimal
a["distance"] = a["distance"].round(2)
```

```python
#Since we don't need latitudes and longitudes in our final  analysis, dropping off those columns
a.drop(['from_latitude', 'from_longitude','to_latitude', 'to_longitude'], inplace = True, axis = 1)
```

```python
#Analysing the columns

a.columns
```

```python
# Coverting "Rentals" df to the same datetime format as "Storm" df before merging
dts = [datetime.datetime.strptime(elem, '%Y-%m-%d %H:%M:%S') 
       for elem in a["start_time"]]
fmted = [elem.strftime('%Y-%m-%d %H:%M:%S') for elem in dts]

a.insert(0, 'start_time1', fmted)
```

```python
a.head(2)
```

```python
# Creating a copy of the original df
q = a.copy()
```

```python
# Display the number of rows and columns
q.shape
```

```python
# Display first  observation from the copied df
q.head(1)
```

```python
# Reading the "Storms" dataset
r = pd.read_csv("Pi_storm.csv", names = ["start_time1", "event_type"], header = 0)
```

```python
#Coverting the common datafields to similar datetime objects
q["start_time1"] =q["start_time1"].astype('datetime64[ns]')
r["start_time1"] =r["start_time1"].astype('datetime64[ns]')
```

```python
#Adding another  temporary Date column to match on these columns before performing pd.asof() merge operation.
q["DATE"] = pd.to_datetime(q['start_time1']).dt.date
```

```python
r["DATE"] = pd.to_datetime(r["start_time1"]).dt.date
```

```python
#Display the first row
r.head(1)
```

```python
"""Performing time series merge using Date & Time by using the common Field name to join on ( on='start_time1' column).
For each row in the left "q" dataFrame, only rows from the right DataFrame ("r") whose 'on' column values match the
 the timedelta value of within tolerance range(1 hour) to  will be kept"""

s = pd.merge_asof(q.sort_values(by ="start_time1") ,r.sort_values(by ="start_time1") , on='start_time1', by ="DATE", tolerance=pd.Timedelta(hours=1)).fillna('NAN')
```
